```python
import pandas as pd
import ast

# Función para convertir la cadena de datos en un diccionario de variables y guardar en un dataframe
def guardar_variables(datos_str):
    datos = ast.literal_eval(datos_str)
    variables = {
        'iva': datos['iva'],
        'is_new': datos['is_new'],
        'is_pack': datos['is_pack'],
        'pack_size': datos['pack_size'],
        'unit_name': datos['unit_name'],
        'unit_size': datos['unit_size'],
        'bulk_price': datos['bulk_price'],
        'unit_price': datos['unit_price'],
        'approx_size': datos['approx_size'],
        'size_format': datos['size_format'],
        'total_units': datos['total_units'],
        'unit_selector': datos['unit_selector'],
        'bunch_selector': datos['bunch_selector'],
        'drained_weight': datos['drained_weight'],
        'selling_method': datos['selling_method'],
        'price_decreased': datos['price_decreased'],
        'reference_price': datos['reference_price'],
        'min_bunch_amount': datos['min_bunch_amount'],
        'reference_format': datos['reference_format'],
        'previous_unit_price': datos['previous_unit_price'],
        'increment_bunch_amount': datos['increment_bunch_amount']
    }
    df = pd.DataFrame.from_dict(variables, orient='index', columns=['value'])
    df.index.name = 'variable'
    return df


# Aplicar la función a cada fila del dataframe
df_resultado = data['price_instructions'].apply(guardar_variables)

# Concatenar todos los dataframes resultantes
df_final = pd.concat(df_resultado.tolist(), keys=df_resultado.index)
```

```python
import pandas as pd
import ast
import numpy as np

# Cargar el DataFrame original
df = pd.read_csv('productos_mercadona_vlc.csv')

# Función para convertir el string a un diccionario
def str_to_dict(s):
    return ast.literal_eval(s)

# Lista de columnas a incluir en el DataFrame final
columnas = ['unit_price', 'bulk_price', 'iva', 'is_new', 'is_pack', 'pack_size', 'unit_name', 'unit_size', 'approx_size', 'size_format', 'total_units', 'unit_selector', 'bunch_selector', 'drained_weight', 'selling_method', 'price_decreased', 'reference_price', 'min_bunch_amount', 'reference_format', 'previous_unit_price', 'increment_bunch_amount']

# DataFrame vacío para guardar los resultados
resultados = []

# Bucle para procesar cada fila del DataFrame original
for index, row in df.iterrows():
    # Convertir la columna 'price_instructions' en un diccionario
    instrucciones = str_to_dict(row['price_instructions'])
    # Crear un diccionario con los resultados
    fila_resultado = {col: instrucciones[col] for col in columnas}
    # Añadir la fila a la lista de resultados
    resultados.append(fila_resultado)

# Convertir la lista de resultados en un DataFrame
resultados = pd.DataFrame(resultados)

# Agregar los resultados al DataFrame original
df = pd.concat([df, resultados], axis=1)

# Eliminar la columna 'price_instructions'
df = df.drop('price_instructions', axis=1)
```
